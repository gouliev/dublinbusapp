```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import os
```

```python
def newFileForEachBusLine(filename, checkAttr='LINEID'):
    with open(filename,newline='') as file:
        reader = csv.DictReader(file, delimiter=';')
        buslines = [] #list will store buslines seen to decide whether to append or not
        for row in reader:
            curLine = row[checkAttr]
            string = ''
            for n in row:
                string += row[n]+',' #adds one comma too many

            if curLine not in buslines:
                writeMethod = 'w'
                buslines.append(curLine)
            else:
                writeMethod = 'a'

            with open('lineSeparated/{}trips'.format(curLine),writeMethod,newline='') as f:
                f.write(string[:-1] +'\n') #last character not used to remove excess comma
```

```python
newFileForEachBusLine("rt_trips_2017_I_DB_1000.txt", checkAttr='lineid')
```
